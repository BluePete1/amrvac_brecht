```python
n = "40"
ds1 = yt.load("maze_128_00"+ n +".dat", unit_system="code")
ds2 = yt.load("maze_adv_128_00" + n + ".dat", unit_system="code")

p1=yt.plot_2d(ds1,"u")
p2=yt.plot_2d(ds2,"u")
p1.set_xlabel("x")
p1.set_ylabel("y")
p2.set_xlabel("x")
p2.set_ylabel("y")
p1.show()
p2.show()

p1.save("GSD" + n + ".png")
p2.save("GSAD" + n + ".png")
```
