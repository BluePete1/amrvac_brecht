```python
import matplotlib

matplotlib.use("Agg")
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import yt

n = "minmod"
ds1 = yt.load("limit_" + n + "_0000.dat", unit_system="code")
ds2 = yt.load("limit_" + n + "_0002.dat", unit_system="code")
ds3 = yt.load("limit_" + n + "_0004.dat", unit_system="code")
ds4 = yt.load("limit_" + n + "_0006.dat", unit_system="code")
ds5 = yt.load("limit_" + n + "_0008.dat", unit_system="code")
ds6 = yt.load("limit_" + n + "_0010.dat", unit_system="code")
_, c1 = ds1.find_max("u")
_, c2 = ds2.find_max("u")
_, c3 = ds3.find_max("u")
_, c4 = ds4.find_max("u")
_, c5 = ds5.find_max("u")
_, c6 = ds6.find_max("u")
ax = 0  # take a line cut along the x axis

# cutting through the y0,z0 such that we hit the max density
ray1 = ds1.ortho_ray(ax, (c1[1], c1[2]))
ray2 = ds2.ortho_ray(ax, (c2[1], c2[2]))
ray3 = ds3.ortho_ray(ax, (c3[1], c3[2]))
ray4 = ds4.ortho_ray(ax, (c4[1], c4[2]))
ray5 = ds5.ortho_ray(ax, (c5[1], c5[2]))
ray6 = ds6.ortho_ray(ax, (c6[1], c6[2]))

# Sort the ray values by 'x' so there are no discontinuities
# in the line plot
srt1 = np.argsort(ray1["index", "x"])
srt2 = np.argsort(ray2["index", "x"])
srt3 = np.argsort(ray3["index", "x"])
srt4 = np.argsort(ray4["index", "x"])
srt5 = np.argsort(ray5["index", "x"])
srt6 = np.argsort(ray6["index", "x"])

plt.figure(figsize=(8,20))
plt.subplot(211)
plt.plot(np.array(ray1["index", "x"][srt1]), np.array(ray1["u"][srt1]), 'b', label='t=0')
plt.plot(np.array(ray3["index", "x"][srt3]), np.array(ray3["u"][srt3]), 'r', label='t=0.1')
plt.plot(np.array(ray2["index", "x"][srt2]), np.array(ray2["u"][srt2]), 'g', label='t=0.2')
plt.plot(np.array(ray4["index", "x"][srt4]), np.array(ray4["u"][srt4]), 'y', label='t=0.3')
plt.plot(np.array(ray5["index", "x"][srt5]), np.array(ray5["u"][srt5]), 'k', label='t=0.4')
plt.plot(np.array(ray6["index", "x"][srt6]), np.array(ray6["u"][srt6]), 'c', label='t=0.5')
plt.xlim(0, 1)
plt.xlabel("x")
plt.ylabel("u")
plt.legend(loc='upper left')

#plt.savefig("begin.png")
#plt.savefig("limiters.png")
```

```python
import matplotlib
matplotlib.use("Agg")
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import yt


n = "0020"
var = "u"
ds1 = yt.load("limit_minmod_" + n + ".dat", unit_system="code")
ds2 = yt.load("limit_kaka_" + n + ".dat", unit_system="code")
_, c1 = ds1.find_max(var)
_, c2 = ds2.find_max(var)
ax = 0  # take a line cut along the x axis

# cutting through the y0,z0 such that we hit the max density
ray1 = ds1.ortho_ray(ax, (c1[1], c1[2]))
ray2 = ds2.ortho_ray(ax, (c2[1], c2[2]))

# Sort the ray values by 'x' so there are no discontinuities
# in the line plot
srt1 = np.argsort(ray1["index", "x"])
srt2 = np.argsort(ray2["index", "x"])

plt.figure(figsize=(8,20))
plt.subplot(211)
plt.plot(np.array(ray1["index", "x"][srt1]), np.array(ray1[var][srt1]), 'b')
plt.plot(np.array(ray2["index", "x"][srt2]), np.array(ray2[var][srt2]), 'r')
plt.xlim(0.35, 0.45)

u1 = np.array(ray1[var][srt1])
u2 = np.array(ray2[var][srt2])

print(100*sum(abs(u1-u2))/sum(abs(u2)))
print(100*sum((u1-u2)**2)**0.5/sum((u2)**2)**0.5)



```
